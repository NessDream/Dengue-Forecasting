```python
companies_list_all = []
for i in range(10):
    url = 'https://www.indeed.fr/jobs?q=data+scientist&l=Paris&start='+str(i*10)
    r = requests.get(url)
    data = r.text
    soup = BeautifulSoup(data, 'html.parser')
    companies_span_list = soup.find_all('span', class_="company")
    companies_list = [company_span.text.replace('\n', '') for company_span in companies_span_list]
    companies_list_all += companies_list
```

```python
companies_list_all.sort()
companies_list_all[:10]
```

```python
companies = list(set(companies_list_all))
companies.sort()
companies[:10]
```
