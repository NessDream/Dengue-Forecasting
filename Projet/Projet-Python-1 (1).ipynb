La dengue est une maladie transmise par les moustiques qui survient dans les régions tropicales et subtropicales du monde. Dans les cas bénins, les symptômes sont similaires à ceux de la grippe: fièvre, éruption cutanée ou encore douleurs musculaires et articulaires. Dans les cas graves, la dengue peut provoquer des saignements graves, une hypotension artérielle et même la mort.

Parce qu'elle est portée par les moustiques, la dynamique de transmission de la dengue est liée à des variables climatiques telles que la température et les précipitations. Bien que la relation avec le climat soit complexe, un nombre croissant de scientifiques soutiennent que le changement climatique est susceptible de produire des changements de distribution qui auront des implications importantes pour la santé publique dans le monde entier.

Ces dernières années, la dengue s'est propagée. Historiquement, la maladie a été la plus répandue en Asie du Sud-Est et dans les îles du Pacifique. De nos jours, la plupart des près de 500 millions de cas par an se produisent en Amérique latine.

Dans ces zones, les schémas de transmission saisonniers sont ponctués par des épidémies beaucoup plus importantes à un interval de quelques années. Étant donné l'imprévisibilité des ces épidémies et leurs conséquences majeures pour les populations touchées, l'objectif de ce projet est de déterminer quels en sont les facteurs clés.

# Description des données

Les données sont des données historiques de surveillance pour deux sites endémiques: Iquitos au Pérou et San Juan à Porto Rico. Elles se composent :

- du nombre de cas hebdomadaires de dengue confirmés en laboratoire depuis l'année 2000 pour Iquitos et l'année 1990 pour San Juan (label)
- de nombreuses variables liées aux conditions météorologiques et à la végétation (features)

Elles proviennent du site web de l'administrationn nationale des océans et de l'atmosphère :


https://dengueforecasting.noaa.gov/

#  Les features


Les features se composent d'une vingtaine de variables :

-Les variables **year** et **weekofyear** représentent respectivement l'année et le numéro de la semaine de cette même année pendant laquelle le nombre de cas de dengue et les attributs météorologiques sont observés.

-La variable **precipitation_amt_mm**  réprésente la somme des précipitations quotidiennes avec pour unité le millimétre (mm).

-Les variables du type "*ndvi_xx*" représentent l'indice de végétation par différence normalisée. Il est est compris entre -1 et 1 :

- Lorsque sa valeur est négative, il est fort probable que la surface terrestre soit de l'eau ou de la neige.
- Avec une valeur proche de 1, il est fort possible que ce soit des feuilles denses.
- Avec une valeur proche de 0, il s'agit de sols nus (zone urbaine par exemple).

Il existe quatre variable de ce type : une pour chaque orientation (sud-est, sud-ouest, nord-ouest et nord-est). On a donc les variables suivantes :
- **ndvi_se**
- **ndvi_sw**
- **ndvi_ne**
- **ndvi_nw**.

-Les variables du type "*station_xxx*"  sont basées sur une estimation ponctuelle des conditions climatiques locales. Elles incluent : 

- La température maximale (**station_max_temp_c**).  
- La température minimale (**station_min_temp_c**).
- La température moyenne quotidienne (**station_avg_temp_c**).
- La variation de températures diurne, i.e. la différence entre la température quotidienne maximale et minimale (**station_diur_temp_rng_c**).
- Les précipitations quotidiennes (**station_precip_mm** qui est équivalent à la variable **precipitation_amt_mm**).\
Les températures sont en degrés Celsius et les précipitations en millimétre (mm).

-Les variables du type "*reanalysis_xx*" sont des données de réanalyse. Elles consituent une bonne représentation des conditions météorologiques d'une zone donnée. En effet, elles sont produites en combinant des mesures, des observations et des données de simulation historiques à l'aide de techniques d'assimilation de données, pour arriver à une description la plus réaliste des phénomènes méteorologiques.

Elles fournissent donc une variété de données qui ne sont pas facilement accessibles à partir d'autres sources comme par exemple l'humidité relative (**reanalysis_relative_humidity_percent**), l'humidité spécifique (**reanalysis_specific_humidity_g_per_kg**) et le point de rosée (**reanalysis_dew_point_temp_k**). Les valeurs des données de température sont en Kelvin.

### Création de jeu de test et d'entraînement :

Pour test et train, on coupe dans le temps, on prend²

```python
years_sj = df[df['city']==1.0]['year'].astype('category').cat.categories.tolist()
years_iq = df[df['city']==0.0]['year'].astype('category').cat.categories.tolist()
```

```python
n_iq = 7 #7 premières années sur 11
n_sj = 14 #14 premières années sur 19

#Train
train_iq = df[(df['city']==0.0) & ( df['year'].isin(years_iq[:n_iq]))]
train_sj = df[(df['city']==1.0) & ( df['year'].isin(years_sj[:n_sj]))]
X_train = pd.concat([train_iq, train_sj]).drop(['total_cases'], axis=1)
y_train = pd.concat([train_iq, train_sj])[['total_cases']]

#Test
test_iq = df[(df['city']==0.0) & ( df['year'].isin(years_iq[n_iq:]))]
test_sj = df[(df['city']==1.0) & ( df['year'].isin(years_sj[n_sj:]))]
X_test = pd.concat([test_iq, test_sj]).drop(['total_cases'], axis=1)
y_test = pd.concat([test_iq, test_sj])[['total_cases']]
```

```python
importances = regr.feature_importances_

# Sort feature importances in descending order
indices = np.argsort(importances)[::-1]

# Rearrange features names so they match the sorted feature importances
names = [list(X_train)[i] for i in indices]

# Create plot
plt.figure()
plt.title("Feature Importance")
plt.barh(range(X_train.shape[1]), importances[indices], color="r", align="center")
plt.yticks(range(X_train.shape[1]), names)

# Show plot
plt.show()
```

# Score

La métrique demandée par le site est l'erreur absolue, c'est donc celle qu'on va utiliser

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import accuracy_score
pred = regr.predict(X_test)
pred = [round(x) for x in pred]

#On pondère par la taille de la soumission pour avoir une idée de notre futur score :
print("accuracy : ", accuracy_score(y_test, pred))
print("\nMAE (DrivenData) : ", round(mean_absolute_error(y_test, pred)*(len(features_test) / len(y_test)), 3))
```
