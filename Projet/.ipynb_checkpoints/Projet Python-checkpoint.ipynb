# Projet Dengue

Projet issu d'une compétition de [DrivenData](https://www.drivendata.org/competitions/44/dengai-predicting-disease-spread/)

La dengue est une maladie transmise par les moustiques qui survient dans les régions tropicales et subtropicales du monde. Dans les cas bénins, les symptômes sont similaires à ceux de la grippe: fièvre, éruption cutanée ou encore douleurs musculaires et articulaires. Dans les cas graves, la dengue peut provoquer des saignements graves, une hypotension artérielle et même la mort.

Parce qu'elle est portée par les moustiques, la dynamique de transmission de la dengue est liée à des variables climatiques telles que la température et les précipitations. Bien que la relation avec le climat soit complexe, un nombre croissant de scientifiques soutiennent que le changement climatique est susceptible de produire des changements de distribution qui auront des implications importantes pour la santé publique dans le monde entier.

Ces dernières années, la dengue s'est propagée. Historiquement, la maladie a été la plus répandue en Asie du Sud-Est et dans les îles du Pacifique. De nos jours, la plupart des près de 500 millions de cas par an se produisent en Amérique latine.

Dans ces zones, les schémas de transmission saisonniers sont ponctués par des épidémies beaucoup plus importantes à un interval de quelques années. Étant donné l'imprévisibilité des ces épidémies et leurs conséquences majeures pour les populations touchées, l'objectif de ce projet est de déterminer quels en sont les facteurs clés.

Est-il possible de prédire l'évolution d'une maladie comme la dengue, en se basant sur des données météo et démographiques ?

Nous étudierons donc les moustiques et les facteurs météorologiques leur permettant de proliférer, ainsi que les données démographiques qui favorisent sa propagation dans la population.

https://www.healthmap.org/dengue/en/index.php

Comme nous avons pu le lire dans diverses sources d'information, la dengue se propage uniquement par les moustiques. Autrement dit, un humain ne peut pas transmettre sa maladie à un autre humain. Par contre, si un moustique peut infecter un humain, l'inverse est vrai également. On a donc un double effet :   
- une météo favorable permet la prolifération des moustiques, et notamment ceux porteurs du virus
- ceux-ci touchent encore plus d'humains si la densité de population est imprtante a priori.
- En retour, plus d'humains infectés provoque une recrudescence du virus chez les moustiques. Ceux-ci vont alors répandre plus encore la maladie.  

Notre modélisation devra donc prendre en compte les données déjà fournies de météo, afin d'identifier clairement quels facteurs provoquent une prolifération des moustiques. On s'attend à voir apparaître surtout la température et le ndvi (indice de végétation par différence normalisée), indice qui révèle une surface d'eau (vers -1) où les moustiques pondent leurs oeufs, de végétation (vers +1) ou de sol nu (vers 0).  

Également nous avons prévu d'y ajouter des données de population pour chacune des deux villes présente dans le dataset.

# Description des données

Les données sont des données historiques de surveillance pour deux sites endémiques: Iquitos au Pérou et San Juan à Porto Rico. Elles se composent :

- du nombre de cas hebdomadaires de dengue confirmés en laboratoire depuis l'année 2000 pour Iquitos et l'année 1990 pour San Juan (label)
- de nombreuses variables liées aux conditions météorologiques et à la végétation (features)

Elles proviennent du site web de l'administration nationale des océans et de l'atmosphère :


https://dengueforecasting.noaa.gov/

#  Les features


Les features se composent d'une vingtaine de variables :

-Les variables **year** et **weekofyear** représentent respectivement l'année et le numéro de la semaine de cette même année pendant laquelle le nombre de cas de dengue et les attributs météorologiques sont observés.

-La variable **precipitation_amt_mm**  réprésente la somme des précipitations quotidiennes avec pour unité le millimétre (mm).

-Les variables du type "*ndvi_xx*" représentent l'indice de végétation par différence normalisée. Il est est compris entre -1 et 1 :

- Lorsque sa valeur est négative, il est fort probable que la surface terrestre soit de l'eau ou de la neige.
- Avec une valeur proche de 1, il est fort possible que ce soit des feuilles denses.
- Avec une valeur proche de 0, il s'agit de sols nus (zone urbaine par exemple).

Il existe quatre variable de ce type : une pour chaque orientation (sud-est, sud-ouest, nord-ouest et nord-est). On a donc les variables suivantes :
- **ndvi_se**
- **ndvi_sw**
- **ndvi_ne**
- **ndvi_nw**.

-Les variables du type "*station_xxx*"  sont basées sur une estimation ponctuelle des conditions climatiques locales. Elles incluent : 

- La température maximale (**station_max_temp_c**).  
- La température minimale (**station_min_temp_c**).
- La température moyenne quotidienne (**station_avg_temp_c**).
- La variation de températures diurne, i.e. la différence entre la température quotidienne maximale et minimale (**station_diur_temp_rng_c**).
- Les précipitations quotidiennes (**station_precip_mm** qui est équivalent à la variable **precipitation_amt_mm**).\
Les températures sont en degrés Celsius et les précipitations en millimétre (mm).

-Les variables du type "*reanalysis_xx*" sont des données de réanalyse. Elles consituent une bonne représentation des conditions météorologiques d'une zone donnée. En effet, elles sont produites en combinant des mesures, des observations et des données de simulation historiques à l'aide de techniques d'assimilation de données, pour arriver à une description la plus réaliste des phénomènes méteorologiques.

Elles fournissent donc une variété de données qui ne sont pas facilement accessibles à partir d'autres sources comme par exemple l'humidité relative (**reanalysis_relative_humidity_percent**), l'humidité spécifique (**reanalysis_specific_humidity_g_per_kg**) et le point de rosée (**reanalysis_dew_point_temp_k**). Les valeurs des données de température sont en Kelvin.

<div class="alert alert-info">
Pour test et train, on coupe dans le temps :   
On prend les premières années pour le train, et on garde les suivantes pour le test.  
Ainsi, on se met dans les mêmes conditions que pour la vériable soumission de résultats.

```python
years_sj = df[df['city']==1.0]['year'].astype('category').cat.categories.tolist()
years_iq = df[df['city']==0.0]['year'].astype('category').cat.categories.tolist()
```

```python
n_iq = 7 #7 premières années sur 12
n_sj = 14 #14 premières années sur 19

#Train
train_iq = df[(df['city']==0.0) & ( df['year'].isin(years_iq[:n_iq]))]
train_sj = df[(df['city']==1.0) & ( df['year'].isin(years_sj[:n_sj]))]
X_train = pd.concat([train_iq, train_sj]).drop(['total_cases'], axis=1)
y_train = pd.concat([train_iq, train_sj])[['total_cases']]

#Test
test_iq = df[(df['city']==0.0) & ( df['year'].isin(years_iq[n_iq:]))]
test_sj = df[(df['city']==1.0) & ( df['year'].isin(years_sj[n_sj:]))]
X_test = pd.concat([test_iq, test_sj]).drop(['total_cases'], axis=1)
y_test = pd.concat([test_iq, test_sj])[['total_cases']]
```

```python
X_test_copy = X_test.copy() #On garde la week_start_date pour la visualisation des résultats
X_train = X_train.drop(['week_start_date'], axis=1)
X_test = X_test.drop(['week_start_date'], axis=1)
```

<div class="alert alert-info">
On **supprime year** pour éviter l'overfit.   
L'année n'est pas censée apporter des informations, seul le contexte (météo, densité etc) doit être inclus dans le modèle pour correctement interpréter les causes des épidémies.

```python
X_train = X_train.drop(['year'], axis=1)
X_test = X_test.drop(['year'], axis=1)
```

# Comparaison de différents Modèles

On teste de nombreux modèles différents, pour voir lesquels seraient le plus adaptés. On s'attend tout de même à une bonne réussite des algorithmes d'ensemble (forêts et boosting de gradient), qui permettront d'identifier avec une plus grande précision les deux villes distinctes de notre dataset, et donc de créer deux modèles assez distincts.  
Ceci est beaucoup plus difficile avec une régression linéaire par exemple, qui est assez rigide, et aura certainement du mal a bien dégager les deux comportements.

```python
def test_model(model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_pred = np.rint(y_pred)
    score = round(mean_absolute_error(y_test, y_pred), 3)
    return(score)
```

```python
lg = LinearRegression()
ridge = linear_model.Ridge()
lasso = linear_model.Lasso(alpha=0.03)
knn = neighbors.KNeighborsRegressor(n_neighbors = 9)
svr = SVR()
rf = RandomForestRegressor(100)
xgb = xgboost.XGBRegressor(n_estimator = 200, silent = True)

score_lg = test_model(lg, X_train, y_train, X_test, y_test)
score_ridge = test_model(ridge, X_train, y_train, X_test, y_test)
score_lasso = test_model(lasso, X_train, y_train, X_test, y_test)
score_knn = test_model(knn, X_train, y_train, X_test, y_test)
score_rf = test_model(rf, X_train, y_train, X_test, y_test)
score_svr = test_model(svr, X_train, y_train, X_test, y_test)
score_xgb = test_model(xgb, X_train, y_train, X_test, y_test)

importances_rf = np.asarray(rf.feature_importances_)
importances_xgb = np.asarray(xgb.feature_importances_)

print([score_lg, score_ridge, score_lasso, score_knn, score_rf, score_svr, score_xgb])
```

On remarque que le prédicteur random (Dummy

# Choix de XGBoost

Comme on peut le voir, l'algorithme le plus précis est XGBoost, comme nous l'avions envisagé.   
C'est donc naturellemnt celui que nous choisront pour la suite.

```python
xgb = xgboost.XGBRegressor(n_estimator = 200, eval_metric="mae", silent = True)

xgb.fit(X_train, y_train)
y_pred = xgb.predict(X_test)
y_pred = np.rint(y_pred)

score = round(mean_absolute_error(y_test, y_pred), 3)
score
```

```python
importances = xgb.feature_importances_

# Sort feature importances in descending order
indices = np.argsort(importances)[::-1]

# Rearrange features names so they match the sorted feature importances
names = [list(X_train)[i] for i in indices]

# Create plot
plt.figure()
plt.title("Xtreme Gradient Boosting")
plt.barh(range(X_train.shape[1]), importances[indices], color="r", align="center")
plt.yticks(range(X_train.shape[1]), names)

# Show plot
plt.show()
```

XGBoost a sélectionné les features pour donner le maximum de poids à celles qui sont le plus déterminantes pour expliquer le taux de nouveaux cas de dangue.  
On a ainsi :  
- **population** : premier facteur de décision, avec 25% de l'explication de la décision. On remarque que la densité est également un facteur important, 3ème avec 7% d'importance. Rajouter ces données nous a dans les faits ramenés d'une mae de 17.5 à la nouvelle, 12.005. C'est donc le point fort de notre analyse.
- **reanalysis_min_air_temp_k** : 2è me variable la plus explicative, avec 18% d'importance. Correspond au minimum de temperature observé par la station météo pendant la semaine. On peut penser que les moustiques effectuent plus de sortie lorsqu'il fait chaud, et par conséquent répandent d'autant plus le virus. Ceci est confirmé par la variable station_avg_temp_c et station_max_temp_c respectivement 5 et 6ème variables, totalisant 11% de poids de décision. La Température est donc un faceur déterminant, juste après les mesures de population.
- **weekofyear** : 4ème variable, 7% de poids. Cette variable semble indiquer un certain effet saisonnier. C'est cohérent avec l'activité des moustiques, bien plus importante en été, et avec les températures qui sont comme on l'a vu importante pour expliquer les nouveaux cas de dengue.
- **eau** : les varaibles d'eau sont en revanche assez faibles comparées à celles citées précédemment. Mais elles ont tout de même une petite participation au modèle, de l'ordre de 15% en totalisant les différentes mesures (plan d'eau (NDVI), pluie, humidité de l'air etc). C'est plutôt une surprise, surtout par rapport à NDVI, qui mesure en quelque sorte la surface "habitable" par les moustiques. Peut-être un certain seuil d'eau permet ensuite aux moustiques de se développer autant qu'il le peuvent, et que leur croissance ne suit pas de façon linéaire leur apport en eau.

### Limites du modèle

Le point essentiel que nous n'avons pas réussi à traiter est d'utiliser de l'autocorrélation, comme on peut le voir en visualisant les séries temporelles du nombre de cas.  

En effet, si la semaine dernière il y a eu beaucoup de cas de dengue, alors ces personnes ont pu contaminer des moustiques sains, et donc accélérer le processus de transmission. Nous avons donc eu l'idée de rajouter la valeur de la semaine passée en utlisant un shift de la target, et de l'insérer dans le jeu d'entrainement ( feature nommée "week-1" sur l'image).    
Cependant cette nouvelle feature prenait toute la place dans l'information, et cela produisait du sur-apprentissage. Nous n'avons pas été capables de résoudre ce problème à temps.

- | - | -

On a alors des résultats assez mauvais, notamment car on peut avoir un effet explosif à cause de ce poids considérable apporté à cette nouvelle feature :

On voit sur les résultats qu'on n'a jamais plus de 4 mois de suite avec autant de malades. L'autocorrélation est donc évidemment beaucoup plus précise sur le jeu d'entrainement, mais pas sur le jeu de test, puisqu'elle se repose sur une prédiction pour définir la valeur suivante, et les erreurs s'accummulent très vite, sans pouvoir être compensée par les données actualisées de météo ou de démographie.  
Il semble donc qu'il faille trouver une approche différente pour utiliser l'information du nombre de cas des semaines passées, malheureusement aucune de nos implémentation n'a aboutit.
